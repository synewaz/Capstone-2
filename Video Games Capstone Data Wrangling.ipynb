Since these are sales, and most likely the NaN values are the same accross the board, I can drop the y columns and rename the x columns as the main sales columns

```python
video_games_df.drop(columns = ['NA_Sales_y', 'EU_Sales_y', 'JP_Sales_y', 'Other_Sales_y', 'Global_Sales_y'],
                                    axis=1, inplace = True)
```

```python
video_games_df.columns
```

```python
video_games_df.dtypes
```

I realize that I should drop also all of the y columns as they're just duplicated columns with not a difference in data, all I care about is the critic score, and user score. I will drop the count columns, and keep the x columns and rename all of the x columns

```python
video_games_df.rename(columns={'NA_Sales_x':'NA_ Sales', 'EU_Sales_x': 'EU_Sales', 'JP_Sales_x':'JP_Sales', 
                               'Other_Sales_x': 'Other_Sales', 'Global_Sales_x':'Global_Sales',
                               'Platform_x':'Platform','Genre_x':'Genre', 'Publisher_x':'Publisher'}, inplace= True)
```

```python
video_games_df.columns
```

```python
video_games_df.drop(columns = ['Platform_y','Genre_y', 'Publisher_y'], axis=1, inplace=True)
```

```python
video_games_df.columns
```

```python
video_games_df.drop(columns=['Critic_Count', 'User_Count'],axis=1, inplace=True)
```

```python
video_games_df.columns
```

```python
video_games_df.drop(columns = ['Developer', 'Year_of_Release'], axis=1, inplace=True)
```

```python
video_games_df.head(45)
```

Before removing/replacing NaN values, we can filter by Year >=2000, and Global Sales >= 1 million

```python
video_games_df = video_games_df[video_games_df['Year'] >= 2000]
video_games_df.head(5)
```

```python
video_games_df = video_games_df[video_games_df['Global_Sales'] >= 1.0]
video_games_df.head(45)
```

Now we must remove NaN rows with our critic scores or UserScores

```python

video_games_df.dropna(subset=['User_Score'],inplace=True)
video_games_df.dropna(subset=['Critic_Score'],inplace=True)
```

```python
video_games_df.head(45)
```

Now to remove duplicate names 

```python
video_games_df= video_games_df.reset_index()
```

```python
video_games_df = video_games_df.drop_duplicates(subset=['Name', 'Rank'])
video_games_df_clean= video_games_df.reset_index()
```

```python
video_games_df_clean.head(45)
```

```python
video_games_df_clean.drop(columns = ['index', 'level_0'], axis=1, inplace=True)
```

```python
video_games_df_clean.head(45)
```

Now that we have cleaned the data, removing all duplicates, and values with NaN. We have filtered on specific global sales anything above 1 mililon, we can now move on to our next step to explore and analyze this data to see which Year, Genre, and publisher had the best output, including their critic/user scores as well.
